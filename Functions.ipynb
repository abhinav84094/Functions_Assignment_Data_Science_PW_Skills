1.  What is the difference between a function and a method in Python?

Function
- A function is a block of code that performs a specific task and is defined using def keyword outside of a class.
```
# This is formatted as code
def greet(name):
    return f"Hello, {name}!"

greet("Abhinav")   #this is the way of calling a funciton
```

Method:
- A method is also a function, but it is defined inside a class and belongs to an object (or class).

```
# This is formatted as code
class Person:
    def greet(self, name):
        return f"Hello, {name}! I am a person."

# Create an object
p = Person()

# Call the method
print(p.greet("Abhinav"))

```

- Standard Function (using def) without any parameter
```

- without any parameter but with return statement
```

```

```

```

```

```

```

    def inner():

6. Explain the concept of generators in Python and how they are defined.

Generators
- A generator in Python is a special type of function that is used to generate a sequence of values one at a time, instead of returning them all at once.
- It is a function that has yield keyword inspite of return.
- It also provides iterable over function
- When we call a generator function, it does not execute immediately —
instead, it returns an iterator-like object (called a generator object), which we can loop over or call next() on.
```
# This is formatted as code
def count_up_to(n):
    count = 1
    while count <= n:
        yield count      # pause here and return current value
        count += 1

counter = count_up_to(3)
print(next(counter))  # 1
print(next(counter))  # 2
print(next(counter))  # 3
# next(counter) → StopIteration

```



7.What are the advantages of using generators over regular functions?

Advantages of Generators in Python
- Memory Efficiency (Low Memory Usage)
  - A regular function that returns a list stores all values in memory at once.
  - A generator produces one value at a time, so it doesn’t need to store the whole sequence.

- Lazy Evaluation (On-Demand Computation)
  - Generators produce values only when required — not all at once.

- Infinite or Very Large Sequences
  - Generators can handle infinite or huge data streams, which normal functions can’t (because of memory limits).

- Faster Execution for Iteration
   - Since generators don’t build large data structures, they are generally faster for iterating large datasets or reading files.

9.Explain the purpose and usage of the `map()` function in Python.
- The map() function in Python is used to apply a function to each item in an iterable (like a list or tuple) and returns a map object (iterator) containing the results.
- map() function takes two parameters, first takes a function and second an iterable object where map function iterates and pass that to function.
- Syntax
  - map(function, iterable)
```
# This is formatted as code
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

result = map(square, numbers)
print(result)          # Output: <map object at ...>
print(list(result))    # Convert to list → [1, 4, 9, 16, 25]
```
- map() applied square() to each element in numbers.
- It returns an iterator, so we usually convert it to a list (or loop through it).
```
# This is formatted as code
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)   # [1, 4, 9, 16]
```
- Lambdas are often used with map() for short, one-line functions.



10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

map()

- map() function takes two parameters, first takes a function and second an iterable object where map function iterates and pass that to function.
- Syntax
  - map(function, iterable)
```

```




reduce()
- It also takes two parameters as a map function.
- It apply function cumulatively to reduce sequence
- It returns a single value
```
# This is formatted as code
reduce(lambda x,y:x+y, [1,2,3,4]) # 10
```



filter()
- It also takes two parameters as a map function
- It Keep elements only that meet a condition
```
# This is formatted as code
filter(lambda x: x%2==0, [1,2,3,4]) #[2,4]
```

````python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci(num):
  a , b = 0,1
  for _ in range(num):
    a,b=b,a+b
    yield a


ans = fibonacci(5)

print(next(ans))
print(next(ans))
print(next(ans))
print(next(ans))
print(next(ans))
````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(e):
  ans = 1
  for i in range(1,e+1):
    ans *= 2
    yield ans


for num in power_of_2(8):
    print(num, end=" ")  # Output: 2 4 8 16 32

print()

ans = power_of_2(5)

print(next(ans))
print(next(ans))
print(next(ans))
print(next(ans))
print(next(ans))



````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    """Generator function that yields each line of a file."""
    with open(filename, 'r') as file:   # Open file in read mode
        for line in file:
            yield line.strip()          # Remove newline character

file_gen = read_file_line_by_line('sample.txt')

print(next(file_gen))  # First line
print(next(file_gen))  # Second line
````
